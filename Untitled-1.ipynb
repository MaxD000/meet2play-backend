```python
"""
Scraper volange-event.com — TOUS les lieux avec données complètes
1) Collecte tous les liens /catalogue/xxx sur toutes les pages
2) Visite chaque fiche pour extraire les vraies données

pip install playwright && playwright install chromium
"""

import csv
from playwright.async_api import async_playwright, Page

CATALOGUE_URL = "https://www.volange-event.com/catalogue"
OUTPUT_FILE   = "volange_lieux.csv"
HEADLESS      = True
TIMEOUT       = 20_000

JS_COLLECT_LINKS = r"""
() => {
    const seen = new Set();
    const links = [];
    document.querySelectorAll("a[href*='/catalogue/']").forEach(a => {
        const href = a.href;
        const slug = href.split('/catalogue/')[1];
        if (!slug || slug === '' || seen.has(href)) return;
        seen.add(href);
        links.push(href);
    });
    return links;
}
"""

JS_SCRAPE_DETAIL = r"""
() => {
    const get = sel => { const el = document.querySelector(sel); return el ? el.innerText.trim() : ''; };
    const getAttr = (sel, attr) => { const el = document.querySelector(sel); return el ? (el.getAttribute(attr) || '') : ''; };

    const titre = get('h1') || get('h2') || '';

    // Texte complet de la page (pour extraire localisation, capacité, etc.)
    const body = document.body.innerText;

    // Image principale (hero/cover)
    const imgSelectors = [
        '[class*="hero"] img', '[class*="cover"] img', '[class*="banner"] img',
        '[class*="main"] img', 'main img', 'article img', 'img'
    ];
    let image = '';
    for (const sel of imgSelectors) {
        const img = document.querySelector(sel);
        if (img) { image = img.src || img.dataset.src || ''; if (image) break; }
    }

    // Localisation
    const locSelectors = [
        '[class*="location"]', '[class*="city"]', '[class*="adress"]', '[class*="address"]',
        '[class*="arrondissement"]', '[class*="district"]', '[class*="postal"]', '[class*="ville"]'
    ];
    let localisation = '';
    for (const sel of locSelectors) {
        const el = document.querySelector(sel);
        if (el) { localisation = el.innerText.trim(); if (localisation) break; }
    }

    // Capacité
    const capSelectors = [
        '[class*="capacity"]', '[class*="capacite"]', '[class*="pers"]',
        '[class*="guest"]', '[class*="invit"]', '[class*="person"]'
    ];
    let capacite = '';
    for (const sel of capSelectors) {
        const el = document.querySelector(sel);
        if (el) { capacite = el.innerText.trim(); if (capacite) break; }
    }

    // Prix
    const prixSelectors = [
        '[class*="price"]', '[class*="prix"]', '[class*="tarif"]',
        '[class*="cost"]', '[class*="budget"]', '[class*="rate"]'
    ];
    let prix = '';
    for (const sel of prixSelectors) {
        const el = document.querySelector(sel);
        if (el) { prix = el.innerText.trim(); if (prix) break; }
    }

    // Description
    const descSelectors = [
        '[class*="description"]', '[class*="desc"]', '[class*="text-style"]',
        '[class*="summary"]', '[class*="about"]', 'main p'
    ];
    let description = '';
    for (const sel of descSelectors) {
        const el = document.querySelector(sel);
        if (el) { description = el.innerText.trim(); if (description) break; }
    }

    return { titre, localisation, capacite, prix, description, image };
}
"""


# ── PHASE 1 : collecte de tous les liens ──────────────────────────────────────

async def collect_all_links(page: Page) -> list[str]:
    all_links: set[str] = set()
    visited_pages: set[str] = set()
    page_num = 1

    await page.goto(CATALOGUE_URL, wait_until="domcontentloaded", timeout=TIMEOUT)
    await page.wait_for_load_state("networkidle")

    while True:
        current_url = page.url
        if current_url in visited_pages:
            break
        visited_pages.add(current_url)

        # Scroll pour déclencher le lazy-load
        for _ in range(8):
            await page.evaluate("window.scrollBy(0, window.innerHeight)")
            await page.wait_for_timeout(400)
        await page.wait_for_timeout(1000)

        links = await page.evaluate(JS_COLLECT_LINKS)
        before = len(all_links)
        all_links.update(links)
        print(f"  Page {page_num} — {len(all_links) - before} nouveaux liens (total : {len(all_links)})")

        # Cherche "page suivante" via les numéros de page dans l'URL
        # Essaie de cliquer sur le dernier numéro de page visible pour connaître le max
        if page_num == 1:
            # Découvre le nombre total de pages
            max_page = await page.evaluate(r"""
                () => {
                    const nums = [...document.querySelectorAll('[class*="pagination"] a, [class*="pager"] a')]
                        .map(a => parseInt(a.innerText))
                        .filter(n => !isNaN(n));
                    return nums.length ? Math.max(...nums) : 0;
                }
            """)
            print(f"  → {max_page} pages détectées")
            if max_page > 1:
                for p_num in range(2, max_page + 1):
                    # Cherche le lien avec ce numéro
                    clicked = await page.evaluate(f"""
                        () => {{
                            const links = [...document.querySelectorAll('[class*="pagination"] a, [class*="pager"] a')];
                            const target = links.find(a => a.innerText.trim() === '{p_num}');
                            if (target) {{ target.click(); return true; }}
                            return false;
                        }}
                    """)
                    if not clicked:
                        break
                    await page.wait_for_load_state("networkidle")
                    for _ in range(8):
                        await page.evaluate("window.scrollBy(0, window.innerHeight)")
                        await page.wait_for_timeout(400)
                    await page.wait_for_timeout(1000)
                    links = await page.evaluate(JS_COLLECT_LINKS)
                    before = len(all_links)
                    all_links.update(links)
                    print(f"  Page {p_num} — {len(all_links) - before} nouveaux liens (total : {len(all_links)})")
                break  # pagination gérée manuellement

        # Sinon, cherche un bouton "suivant"
        next_el = await page.query_selector(
            "a[rel='next'], [aria-label='Next'], [aria-label='Suivant'], "
            "[class*='next']:not([class*='wiper'])"
        )
        if not next_el:
            break
        disabled = await next_el.get_attribute("disabled")
        aria_disabled = await next_el.get_attribute("aria-disabled")
        if disabled is not None or aria_disabled == "true":
            break
        await next_el.click()
        await page.wait_for_load_state("networkidle")
        page_num += 1

    return sorted(all_links)


# ── PHASE 2 : scraping de chaque fiche ────────────────────────────────────────

async def scrape_venue(page: Page, url: str) -> dict:
    await page.goto(url, wait_until="domcontentloaded", timeout=TIMEOUT)
    await page.wait_for_load_state("networkidle")
    data = await page.evaluate(JS_SCRAPE_DETAIL)
    data['lien'] = url
    return data


# ── EXPORT ────────────────────────────────────────────────────────────────────

def save_to_csv(venues: list[dict], filepath: str) -> None:
    if not venues:
        print("⚠ Aucun lieu trouvé.")
        return
    fields = ["titre", "localisation", "capacite", "prix", "description", "lien", "image"]
    with open(filepath, "w", newline="", encoding="utf-8-sig") as f:
        writer = csv.DictWriter(f, fieldnames=fields, extrasaction="ignore")
        writer.writeheader()
        writer.writerows(venues)
    print(f"\n✅ {len(venues)} lieux exportés → {filepath}")


# ── MAIN ──────────────────────────────────────────────────────────────────────

async def main():
    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=HEADLESS)
        context = await browser.new_context(
            user_agent=(
                "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                "AppleWebKit/537.36 (KHTML, like Gecko) "
                "Chrome/120.0.0.0 Safari/537.36"
            ),
            locale="fr-FR",
        )
        page = await context.new_page()

        print("🔗 Phase 1 : collecte des liens...\n")
        links = await collect_all_links(page)
        print(f"\n📋 {len(links)} fiches à scraper\n")

        print("📄 Phase 2 : scraping de chaque fiche...\n")
        venues = []
        for i, url in enumerate(links, 1):
            try:
                data = await scrape_venue(page, url)
                venues.append(data)
                print(f"  [{i}/{len(links)}] ✓ {data.get('titre') or url}")
            except Exception as e:
                print(f"  [{i}/{len(links)}] ✗ {url} → {e}")

        await browser.close()

    save_to_csv(venues, OUTPUT_FILE)


await main()
```

```python
"""
Phase 3 — Détection Paris Society via DuckDuckGo (gratuit, sans clé API)
pip install duckduckgo-search
"""

import csv
import time
from duckduckgo_search import DDGS

INPUT_FILE  = "volange_lieux.csv"
OUTPUT_FILE = "volange_lieux.csv"


def check_paris_society(titre: str) -> str:
    if not titre:
        return "Non"

    query = f'"{titre}" "Paris Society"'

    try:
        with DDGS() as ddgs:
            results = list(ddgs.text(query, max_results=5))

        # Vérifie si Paris Society apparaît dans les résultats
        for r in results:
            text = (r.get("title", "") + " " + r.get("body", "") + " " + r.get("href", "")).lower()
            if "paris society" in text and titre.lower()[:6] in text:
                return "Oui"
    except Exception as e:
        print(f"    ⚠ Erreur recherche pour '{titre}' : {e}")

    return "Non"


def enrich():
    with open(INPUT_FILE, encoding="utf-8-sig") as f:
        rows = list(csv.DictReader(f))

    if not rows:
        print("⚠ CSV vide.")
        return

    # Dédoublonne par titre pour ne pas chercher deux fois le même lieu
    seen_titres: dict[str, str] = {}
    print(f"📂 {len(rows)} lignes à traiter\n")

    for i, row in enumerate(rows, 1):
        titre = row.get("titre", "")

        if titre in seen_titres:
            row["paris_society"] = seen_titres[titre]
        else:
            result = check_paris_society(titre)
            seen_titres[titre] = result
            row["paris_society"] = result
            time.sleep(0.8)  # pause pour éviter le rate-limit DuckDuckGo

        label = "✅ OUI" if row["paris_society"] == "Oui" else "❌ Non"
        print(f"  [{i}/{len(rows)}] {label} — {titre or '(sans titre)'}")

    fieldnames = list(rows[0].keys())
    with open(OUTPUT_FILE, "w", newline="", encoding="utf-8-sig") as f:
        writer = csv.DictWriter(f, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(rows)

    oui = sum(1 for r in rows if r.get("paris_society") == "Oui")
    print(f"\n✅ Terminé — {oui} lieux Paris Society sur {len(rows)} → {OUTPUT_FILE}")


enrich()
```
